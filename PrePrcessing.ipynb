# Preprocessing Data Stroke

```python
ds = pd.read_csv("C:/Users/LENOVO/OneDrive/Desktop/SEMINAR HASIL/Code TA/data/healthcare-dataset-stroke-data.csv")
ds
```

```python
ds = ds.drop(['id'], axis=1)
ds
```

```python
ds.dtypes
```

```python
print(ds.gender.unique())
print(ds.ever_married.unique())
print(ds.work_type.unique())
print(ds.Residence_type.unique())
print(ds.smoking_status.unique())
```

```python
ds.gender.value_counts().plot.pie(autopct='%.2f')
```

```python
ds= ds.replace(to_replace = 'Female', value = 0)
ds= ds.replace(to_replace = 'Male', value = 1)
ds= ds.replace(to_replace = 'Other', value = 1)
ds= ds.replace(to_replace = 'Yes', value = 1)
ds= ds.replace(to_replace = 'No', value = 0)
ds
```

```python
ds = pd.get_dummies(data=ds, columns=["work_type"])
ds = pd.get_dummies(data=ds, columns=["Residence_type"])
ds = pd.get_dummies(data=ds, columns=["smoking_status"])
ds
```

```python
ds.isna().sum()
```

```python
rata = ds['bmi'].mean()
ds['bmi'] = ds['bmi'].fillna(rata)

ds.isna().sum()
```

```python
ds.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

col = ['age', 'avg_glucose_level', 'bmi']

scaler=MinMaxScaler()
ds[col] = scaler.fit_transform (ds[col])
ds
```

```python
#ds.to_csv("Data Stroke.csv", index=False)
```

```python
x = ds.drop('stroke', axis=1)
y = ds['stroke']
print (x.shape)
print (y.shape)
```

```python
y.value_counts()
```

```python
y.value_counts().plot.pie(autopct='%.2f')
```
